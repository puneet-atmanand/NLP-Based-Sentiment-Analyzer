```python
import pandas as pd
```

```python
import nltk#nltk stands for natural laanguage tool kit,that helps machine providing some english words to understand the sentiments of the user in a sentence.
from nltk.corpus import stopwords# corpus is like a bag or a group that is consisted with different kinds of english words.
from nltk.stem import PorterStemmer# PortStemmer is a python library that helps the machine to break the words into their root forms.
import string# String is that kind of a library in python that helps machine read each and every word as a token in a sentence.
```

```python
df=pd.read_csv("/content/sentiment_data.csv", encoding='latin1')
```

```python
df
```

```python
nltk.download("stopwords")
stop_words = stopwords.words("english")#It will check for stopwords(Non-relevant words inside a sentence)
ps = PorterStemmer()#It will continue with the stemming process,and will bring all the words to their root forms.It converts all words to their root forms by removing suffixes like ‚Äú-ing‚Äù and ‚Äú-ed‚Äù from the end of words.
```

```python
def preprocessing(text):
    text = text.lower()
    token = nltk.word_tokenize(text)
    new = []
    for i in token:
        if i not in stop_words and i not in string.punctuation:
            stemmer = ps.stem(i)
            new.append(stemmer)
    return " ".join(new)
```

```python
import nltk
nltk.download('punkt_tab')
df["clean"]=df["Comment"].fillna('').apply(preprocessing)
```

```python
df["clean"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer#TFid Stands for term frequency - inverse document frequency.
```

```python
vector = TfidfVectorizer()
X = vector.fit_transform(df["clean"])
y = df["Sentiment"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=55
)
```

```python
log=LogisticRegression()
nan_mask = y_train.isna()
X_train_filtered = X_train[~nan_mask.values]
y_train_filtered = y_train[~nan_mask]
log.fit(X_train_filtered, y_train_filtered)
```

```python
print("train accuracy",log.score(X_train_filtered,y_train_filtered))
nan_mask_test = y_test.isna()
X_test_filtered = X_test[~nan_mask_test.values]
y_test_filtered = y_test[~nan_mask_test]
print("test accuracy",log.score(X_test_filtered,y_test_filtered))
```

```python
def pred_text(text):
  text=text.lower()
  text=preprocessing(text)
  vec=vector.transform([text])
  predict=log.predict(vec)[0]
  if predict==0:
    return "negative üò†"
  elif predict==1:
    return "neutral üòê"
  else: # predict == 2
    return "positive üòä"
```
